### Character Tokenization

```python
text = "Hello, how are you doing today?"

# Character tokenization
char_tokens = list(text)

print("Original text:", text)
print("Tokens:", char_tokens)
print("Number of tokens:", len(char_tokens))
```

```python
# Word tokenization
word_tokens = text.split()

print("Tokens:", word_tokens)
print("Number of tokens:", len(word_tokens))
```

### Subword Tokenization

```python
import tiktoken

# GPT ka tokenizer
enc = tiktoken.get_encoding("cl100k_base")

# Encode karo
token_ids = enc.encode(text)

# Har token ID ko readable string mein badlo
subword_tokens = [enc.decode([t]) for t in token_ids]

print("Token IDs:", token_ids)
print("Tokens:", subword_tokens)
print("Number of tokens:", len(token_ids))
```

```python
print(f"{'Scheme':<12} {'Tokens':<5}")
print("-" * 35)
print(f"{'Character':<12} {len(char_tokens):<5}")
print(f"{'Word':<12} {len(word_tokens):<5}")
print(f"{'Subword':<12} {len(token_ids):<5}")
```

```python
hindi_text = "नमस्ते, आप कैसे हैं?"

# Character
hindi_char = list(hindi_text)

# Word
hindi_word = hindi_text.split()

# Subword
hindi_ids = enc.encode(hindi_text)
hindi_subword = [enc.decode([t]) for t in hindi_ids]

print(f"Text: {hindi_text}")
print(f"\nCharacter tokens: {len(hindi_char)}")
print(f"Word tokens:      {len(hindi_word)}")
print(f"Subword tokens:   {len(hindi_ids)}")
print(f"\nSubword breakdown: {hindi_subword}")
```

## Tokenization in NLP (Natural Language Processing)
### Character Tokenization
This is simple but inefficient. Each character is treated as a separate token, which increases sequence length and makes learning harder for models.

### Word Tokenization
This is better than character-level tokenization since it works with whole words. However, it struggles with punctuation, unknown words, and variations (like “run”, “running”, “ran”).

### Subword Tokenization
This is the method used by real LLMs. It breaks words into smaller meaningful units (subwords), handling unknown words and reducing vocabulary size effectively.

## Difference Between Hindi and English Tokenization
### English
Tokenization is relatively easier because words are usually separated by spaces and have simpler morphology.

### Hindi
Tokenization is more complex due to:
Rich morphology (many word forms)
Compound words
Script (Devanagari) and joining rules
This makes subword tokenization especially important for Hindi.
