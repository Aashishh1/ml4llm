```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import matplotlib.pyplot as plt
import numpy as np

tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")
model.eval()

print("Ready!")
```

```python
text = "The cat sat on the"
inputs = tokenizer(text, return_tensors="pt")

# Method 1: Greedy — hamesha top word choose karo
greedy_output = model.generate(
    **inputs, 
    max_new_tokens=10,
    do_sample=False
)

# Method 2: Sampling — probability ke hisaab se random
sample_output1 = model.generate(
    **inputs, 
    max_new_tokens=10,
    do_sample=True,
    temperature=1.0
)

# Method 3: High temperature — zyada random
sample_output2 = model.generate(
    **inputs, 
    max_new_tokens=10,
    do_sample=True,
    temperature=2.0
)

print(f"Input: {text}\n")
print(f"Greedy:      {tokenizer.decode(greedy_output[0])}")
print(f"Sample 1.0:  {tokenizer.decode(sample_output1[0])}")
print(f"Sample 2.0:  {tokenizer.decode(sample_output2[0])}")
```

```python
temperatures = [0.1, 0.5, 1.0, 1.5, 2.0]

print(f"Input: {text}\n")
print("="*60)

for temp in temperatures:
    output = model.generate(
        **inputs,
        max_new_tokens=10,
        do_sample=True,
        temperature=temp
    )
    generated = tokenizer.decode(output[0])
    print(f"Temp {temp}: {generated}")
    print()
```

* **Greedy decoding** always produces the **same output**, since it picks the highest-probability token every time.

* **Temperature** controls the **creativity** of the model’s output.

* **Low temperature** → more **safe and predictable**, but often **boring**.

* **High temperature** → more **creative and diverse**, but can introduce **errors or randomness**.

* The **sweet spot is around 1.0**, balancing creativity and coherence effectively.
